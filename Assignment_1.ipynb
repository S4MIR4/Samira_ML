```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
#Reading  the dataset
file_name = "c:\@@@from_drive_S@@@\ECE1513\Indian_liver_patient_assignment_1.csv"
df = pd.read_csv(file_name) # data frame
df = df.fillna(0) #update df
df.loc[df['Gender'] == 'Female', 'Gender'] = 1
df.loc[df['Gender'] == 'Male', 'Gender'] = 0
```

```python
df
```

```python
X = df[list(df.columns)[0:10]]
Y = df['Dataset']
```

```python
# random division to 70% training and 30% testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, train_size = 0.7)
```

```python
#Model 1 = Logistic regression model
model_1 = LogisticRegression() #initialize the model
model_1.fit(X_train, Y_train)  #fitting on the x and y train
Y_prediction = model_1.predict(X_test)
y_train_prediction = model_1.predict(X_train)
training_accuracy = accuracy_score(y_train_prediction, Y_train)
Accuracy = accuracy_score(Y_prediction, Y_test)
print(f"the lg accuracy : {round(Accuracy, 3)}")
print(f"the lg traning accuracy : {round(training_accuracy, 3)}")

#model 2 = SVM
model_2 = svm.SVC(kernel='linear') #initialize the model
model_2.fit(X_train, Y_train)  #fitting on the x and y train
Y_prediction = model_2.predict(X_test)
Accuracy = accuracy_score(Y_prediction, Y_test)
print(f"the svm accuracy : {round(Accuracy, 2)}")

#model 3 = Decision Tree
model_3 = DecisionTreeClassifier() #initialize the model
model_3.fit(X_train, Y_train)  #fitting on the x and y train
Y_prediction = model_3.predict(X_test)
Accuracy = accuracy_score(Y_prediction, Y_test)
print(f"the Decision Tree accuracy : {round(Accuracy, 2)}")

#model 4 = Linear Discriminant Analysis
model_4 = LinearDiscriminantAnalysis() #initialize the model
model_4.fit(X_train, Y_train)  #fitting on the x and y train
Y_prediction = model_4.predict(X_test)
Accuracy = accuracy_score(Y_prediction, Y_test)
print(f"the Linear Discriminant Analysis accuracy : {round(Accuracy, 2)}")
```

```python
#model 5 = KNN classifier

k_values = range(1, 21)

accuracy_values = []
for k in k_values:
    model_5 = KNeighborsClassifier(n_neighbors=k) #initialize the model
    model_5.fit(X_train, Y_train)  #fitting on the x and y train
    y_train_prediction = model_5.predict(X_train)
    training_accuracy = accuracy_score(y_train_prediction, Y_train)
    Y_prediction = model_5.predict(X_test)
    Accuracy = accuracy_score(Y_prediction, Y_test)
    #print(f"the kneighbors classifier accuracy for k={k}: {round(Accuracy, 2)}")
    accuracy_values.append(Accuracy)
    #print(f"the training accuracy of  kneighbors classifier accuracy : {round(training_accuracy, 2)}")


```

```python
plt.figure()
plt.plot(k_values, accuracy_values)
plt.xticks(k_values)
```
